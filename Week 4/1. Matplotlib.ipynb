```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##### 2D line plot

```python
hours_1 = np.random.randint(1,5,7)
days_1 = np.arange(1,8)
hours_2 = np.random.randint(1,5,7)
days_2 = np.arange(1,8)
plt.plot(days_1,hours_1)
plt.plot(days_2,hours_2)
plt.xlabel('Days in week')
plt.ylabel('Hours study in a day')
plt.title('Daily Study')
plt.show()
```

##### 2D data from pandas dataframe

```python
df = pd.read_csv('enrollment_data.csv')
```

```python
plt.plot(df['Year'],df['Programming'],label = 'Programming',color = 'r',marker = 'o',linestyle = '-.')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing',marker = '*',linestyle = '--')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

##### Scatterplot

```python
df = pd.read_csv('student_IQdata.csv')
df.head()
```

```python
plt.scatter(df['Study_Hour'],df['IQ_Score'],edgecolors='r',label='IQ Score',marker='*')
plt.xlabel('Study hours')
plt.ylabel('IQ Score')
plt.grid()
plt.legend()
```

```python
plt.scatter(df['Chilling_Hours'],df['IQ_Score'],edgecolors='g',label='Chilling_Hours',marker='o',c='r')
plt.xlabel('Chilling hours')
plt.ylabel('IQ Score')
plt.grid()
plt.legend()
```

##### Histogram

```python
plt.hist(df['IQ_Score'],bins=5,histtype="barstacked",color='g',density=True,edgecolor='red')
```

##### Histogram 2D

```python
arr = np.arange(1,df['IQ_Score'].size+1)
data = pd.Series(arr)
plt.hist2d(df['IQ_Score'],data)
```
