```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
from click import style
from matplotlib.pyplot import xlabel, ylabel
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/phitronio/Python-for-ML/refs/heads/main/student_IQdata.csv')
sns.set_style(style='darkgrid')
sns.set_palette('husl')
df.head()
```

###### Create a scatter plot with Study_Hours on the X-axis and IQ on the Y-axis with seaborn


```python
fig = sns.scatterplot(data=df,x='Study_Hour',y='IQ_Score')
fig.set(xlabel='Study Hour',ylabel='IQ Score')
plt.show()
```

###### Create a scatter plot with Shoe_Size on the X-Axis and Study_Hours on the Y-axis with seabron

```python
sns.scatterplot(data=df,x='Shoe_Size',y='Study_Hour')
plt.show()
```

###### Create a histogram of Shoe_Size using plotly

```python
fig = px.histogram(data_frame=df, x='Shoe_Size')
fig.show()
```
