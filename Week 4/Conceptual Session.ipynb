##### Web Scrapping

```python
head = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

res = requests.get('https://en.wikipedia.org/wiki/List_of_companies_of_Bangladesh',headers=head).text
```

```python
soup = BeautifulSoup(res)
print(soup.prettify())
```

```python
table = soup.find('table')
table
```

```python
data = table.find_all('tr')
print(type(data[3]))
```

```python
data = data[2:]
```

```python
company = []
industry = []
sector = []
headquaters = []
founded = []
notes = []
for i in data:
    td = i.find_all('td')
    company.append(td[0].text.strip())
    industry.append(td[1].text.strip())
    sector.append(td[2].text.strip())
    headquaters.append(td[3].text.strip())
    founded.append(td[4].text.strip())
    notes.append(td[5].text.strip())

data = {'Company':company,'Industry':industry,'Sector':sector,'Headquaters':headquaters,'Founded':founded,'Notes':notes}
```

```python
df = pd.DataFrame(data)
df.head()
```
