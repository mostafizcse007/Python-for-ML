```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

##### Lineplot (Seaborn)

```python
df = pd.read_csv('sns_data (1).csv')
sns.set_theme(style='darkgrid')
df.head()
```

```python
sns.lineplot(data = df, x= 'week', y= 'attendance_rate', errorbar=None, markers='o', hue= 'gender') #axes label
```

```python
#axes label
sns.relplot(kind='line', data = df, x= 'week', y= 'attendance_rate', errorbar=None, markers='o', hue= 'gender')
```

##### Scatter Plot(Seaborn)

```python
sns.relplot(kind='scatter',data=df,x='study_hours',y='test_score',hue='gender',style='hostel',size='gender')
```

```python
df = sns.load_dataset('tips')
sns.relplot(kind='scatter',data=df,x='total_bill',y='tip',hue='sex',size='sex',style='sex')
plt.show()
```

##### Facet Plot(Seaborn)

```python
df = pd.read_csv('student_dataset_complete.csv')
df.head()
```

```python
val = sns.relplot(kind='scatter',data=df,x='study_hours',y='test_score',col='gender',hue='gender',row='hostel')
val.set(xlabel='Study Hours',ylabel='Test Score')
plt.show()
```

##### Histogram

```python
# axes level
sns.histplot(data=df,x='test_score',hue='gender',bins=10,element='step')
```

```python
sns.histplot(data=df,x='attendance_rate',y='class_name')
```

```python
# figure level
sns.displot(kind='hist', data=df,x='test_score',hue='gender',bins=10,element='step',col='gender')
```

##### KDE Plot(Seaborn)

```python
# kde = kernel density estimation
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(kind='kde',data = tips,x='total_bill',hue='sex',fill=True)
```

##### Count & Bar Plot(Seaborn)

```python
sns.countplot(data=tips,x='sex',hue='time')
```

```python
sns.barplot(data=tips,x='day',y='smoker',hue='sex',errorbar=None)
```

##### Reg Plot(Seaborn)

```python
# reg = regression plot
df = pd.read_csv('student_dataset_complete.csv')
df.head()
# axex level
sns.regplot(data=df,x='time_study',y='test_score',color='r',marker='x')
```

```python
# figure level
sns.lmplot(data=df,x='time_study',y='test_score',hue='gender')
```

##### Pair & Joint Plot

```python
df.head()
sns.pairplot(data=df[['Marks','study_hours','attendance_rate','gender']],kind='reg')
```

```python
sns.jointplot(data=df,x='study_hours',y='Marks')
```

```python
sns.jointplot(data=df,x='study_hours',y='Marks',kind='kde')
```

##### Scatter Plot (Plotly)

```python
fig = px.scatter(data_frame=df,x='time_study',y='Marks',title='ABC',subtitle='XYZ',facet_col='gender',color='gender'
                 ,size='Tshirt_size',hover_data=['student_id','class_name'])
fig.show()
```

##### Line Plot(Plotly)

```python
df = pd.DataFrame({'x':[x for x in range(50)],
                  'y':[x**2 for x in range(50)]})
fig = px.line(data_frame=df,x='x',y='y',markers=True)
fig.show()
```

```python
fig.write_html('html_file.html')
fig.write_json('img.json')
fig.write_image('image.png')
```
