```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('student_IQdata.csv')
df.head(1)
```

###### Create a scatter plot with Study_Hours on the X-axis and IQ on the Y-axis.


```python
plt.scatter(x=df['Study_Hour'], y=df['IQ_Score'],edgecolors='green',label='IQ vs Study')
plt.xlabel('Study Hour')
plt.xlabel('Student IQ')
plt.title('Study Hour vs IQ')
plt.legend()
plt.plot()
```

###### Create a scatter plot with Shoe_Size on the X-Axis and Study_Hours on the Y-axis

```python
plt.scatter(x=df['Shoe_Size'], y=df['Study_Hour'],edgecolors='green',label='Shoe vs Study')
plt.xlabel('Shoe Size')
plt.xlabel('Study Hour')
plt.title('Shoe Size vs Study Hour')
plt.legend()
plt.plot()
```

###### Create a histogram of Shoe_Size

```python
plt.hist(x=df['Shoe_Size'], color='green',edgecolor='black',bins=5)
plt.xlabel('Shoe Size')
plt.ylabel('Count')
plt.show()
```
