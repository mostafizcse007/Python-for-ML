```python
import requests
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/phitronio/Python-for-ML/refs/heads/main/Practice%20Day%20(1).csv'
res = requests.get(url)
df = pd.read_csv(StringIO(res.text),on_bad_lines='skip')
df.head()
```

```python
df['Discounted_Price'] = df['Price'] * 0.90
df.head()
```

```python
df.drop_duplicates(subset=['EmployeeID'],inplace=True)
```

```python

df.dropna(subset='Salary',inplace=True)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df[['Age','Salary']].isnull().sum()
```

```python
plt.hist(df['Age'])
plt.show()
```

```python
plt.scatter(df['Age'],df['Salary'])
plt.show()
```

```python
plt.scatter(df['Price'],df['Discounted_Price'])
plt.show()
```
