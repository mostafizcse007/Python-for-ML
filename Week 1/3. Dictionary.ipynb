```python
def histogram(string):
    d = dict()
    for char in string:
        if char not in d:
            d[char] = 1
        else:
            d[char]+=1
    return d
txt = 'apple'
print(histogram(txt))
```

```python
def cnt_nums(lst):
    d = dict()
    d['pos'] = 0
    d['neg'] = 0
    for num in lst:
        if num < 0:
            d['neg'] += 1

        else:
            d['pos'] += 1
    return d
lst = list(range(-5,19))
print(cnt_nums(lst))
```
