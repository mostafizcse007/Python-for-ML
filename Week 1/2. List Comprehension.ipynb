###### Problem: Given nums = [1,2,3,4,5], make a list of squares.

```python
nums = [x for x in range(21) if x % 2 == 0]
nums
```

###### Problem: From words = ["cat","house","elephant","rat"], keep only words with length > 3

```python
words = ["cat","house","elephant","rat"]
new_words = [item for item in words if len(item)>3]
new_words
```

###### Problem: names = ["bob","Anna","BOB","alice"] → produce unique uppercase names

```python
names = ["bob","Anna","BOB","alice"]
unique_names = {x.capitalize() for x in names}
unique_names
```

###### Problem: letters = ['a','b','c'] → [(1,'a'), (2,'b'), (3,'c')] (1-based index)

```python
letters = ['a','b','c']
new_letters = [(i,ch) for i,ch in enumerate(letters,start=1)]
new_letters
```

###### Problem: prices = [99.99, 120.0, 35.5] → add 10% tax and round to 2 decimals

```python
prices = [99.99, 120.0, 35.5]
with_tax = [round(x+(x*0.1),2) for x in prices]
with_tax
```

###### Problem: matrix = [[1,2,3],[4,5],[6]] → flatten to [1,2,3,4,5,6]

```python
matrix = [[1,2,3],[4,5],[6]]
new_matrix = [val for row in matrix for val in row]
new_matrix
```
