###### Create a Pandas Series from the list [10, 20, 30, 40, 50]. Print the Series, its index, and data type. Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data.

```python
import pandas as pd
```

```python
lst = [10, 20, 30, 40, 50]
df = pd.Series(lst)
print(df)
print(df.dtypes)
```

```python
data = {
    "Name":['A','B','C','D'],
    "Age":[1,2,3,4],
    "City":['AB','BC','CD','AD']
}
df = pd.DataFrame(data)
print(df)
```

###### Create a DataFrame using the dictionary below:


data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}

Print .columns, .index. Access a single column and one full row.

```python
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}
df = pd.DataFrame(data)
print(df.columns,df.index)
print(df.iloc[0:1,0])
```

###### Using .loc[] and .iloc[], show the first row, the second column, and a 2Ã—2 block    from the middle of your DataFrame.

```python
print(df.loc[0:1,:])
print(df.loc[:,'Price'])
print(df.iloc[1:3,0:2])
```
