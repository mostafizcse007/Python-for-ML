#####  1. Show employees with Salary > 45000.
   2. Show employees from the IT department only.
   3. Show employees with Salary > 45000 and Experience > 3.
   4. Sort the employees by Experience (ascending).

```python
data = {
 	'Name': ['Rafiq', 'Nadia', 'Tuhin', 'Rima', 'Faruk'],
 	'Department': ['HR', 'IT', 'IT', 'Sales', 'Sales'],
 	'Salary': [40000, 55000, 52000, 45000, 47000],
 	'Experience': [2, 5, 4, 3, 6]
 }
df = pd.DataFrame(data)
print('IT departemnt only',df[df['Department']=='IT'])
print('Salary > 45000 and Experience > 3',df[(df['Salary']>45000) & (df['Experience']>3)])
print('Sort by experience\n',df.sort_values('Experience',axis = 0))
```

#####  1. Show only the cities where FeelsLike > 33.
 2. Sort the DataFrame by FeelsLike (descending).
 3. Find which city has the highest humidity.


```python
data = {
 	'City': ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna'],
 	'Temp': [32, 28, 30, 35, 33],
 	'Humidity': [60, 75, 70, 55, 65]
 }
df = pd.DataFrame(data)
df.head()
print(df.sort_values('Humidity',ascending=False))
print(df[df['Humidity']==df['Humidity'].max()])
df.loc[df['Humidity'].idxmax(),'City']
```

```python
data = {'income':[20000,30000,50000,90000,124738,4534]}
accounts = pd.DataFrame(data)
mpp = {'Low Salary':0,'Average Salary':0,'High Salary':0}
for i in accounts['income']:
    if i < 20000:
        mpp['Low Salary'] += 1
    elif 20000<= i <= 50000:
        mpp['Average Salary'] += 1
    else:
        mpp['High Salary'] += 1

df = pd.DataFrame(list(mpp.items()),columns=['category','accounts_count'])
print(df)
```
