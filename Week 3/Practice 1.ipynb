```python
import numpy as np
```

##### Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.


```python
one_D = np.arange(10)
two_D = np.arange(1,11).reshape(2,5)
three_D = np.arange(1,28).reshape(3,3,3)
print('1D\n',one_D)
print('2D\n',two_D)
print('3D\n',three_D)
```

##### an array of zeros of shape (3,3)


```python
arr = np.zeros(shape=(3,3),dtype=np.int8)
arr
```

##### an array of ones of shape (2,5)


```python
arr = np.ones(shape=(2,5),dtype=np.int8)
arr
```

##### an array filled with 7

```python
arr = np.full((3,3),7)
arr
```

#####  Create an array of even numbers between 10 and 50 using arange

```python
arr = np.arange(10,51,2)
arr
```

##### Create a 4×4 matrix from 0–15

```python
arr = np.arange(0,16).reshape(4,4)
print('First row\n',arr[0:1,:])
print('Last row\n',arr[:,3])
print('2 x 2 subarray from the center\n',arr[1:3,1:3])
```

#####  Add a 1D array [1,2,3] to each row of a 3×3 matrix.


```python
arr = np.arange(1,10).reshape(3,3)
brr = np.array([1,2,3])
print(arr+brr)
```
