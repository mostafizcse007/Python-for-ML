###### Sorting

```python
x = np.random.randint(10,20,(10,))
x.sort() # inplace sorting
print(x)

y = np.random.randint(10,20,(10,))
copied = np.sort(y)
print(copied)
print(y)
```

##### 2D array sort

```python
arr = np.arange(25,0,-1).reshape(5,5)
hor_sort = np.sort(arr, axis= 1) # row wise
print(hor_sort)
ver_sort = np.sort(arr, axis = 0,kind='heapsort')
print(ver_sort)
print(arr)
```

##### Searching

```python
arr = np.array([1,2,6,2,5,8,5,2,1,5,8,998,43,7])
ind = np.where(arr == 5, arr+65, arr)
print(ind)
print(np.argmax(arr)) # maximum value index
print(np.argmin(arr)) # minimum value index
```

##### Counting

```python
arr = np.random.randint(0,100,size=(5,10))
print(np.count_nonzero(arr>50))
print(np.unique(arr,return_counts=True))
```

##### Statistics

```python
data = np.genfromtxt('students_marks.csv',delimiter=',')
data = data[1:,1:]
print("Science",data[:,0:1].max())
print("Math",data[:,1:2].max())
print("English",data[:,2:].max())
```

```python
study_hr = np.array([2,3,4,5,6,7,8])
marks = np.array([50,55,60,70,75,80,90])
data = np.array([study_hr,marks])
print("Correlation Matrix \n",np.corrcoef(data))
```
