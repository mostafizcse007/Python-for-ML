##### Question 5

```python
def factorial(n):
    if n==0 or n == 1:
        return 1
    return n * factorial(n-1)

num = int(input())
copy = num
sum = 0
while num:
    temp = (num % 10)
    sum += factorial(temp)
    num //= 10
if copy == sum:
    print("Strong Number")
else:
    print("Not Strong Number")
```

##### Question 3

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self,w,h):
        self.w = w
        self.h = h

    def area(self):
        return float(self.w * self.h)

class Circle(Shape):
    def __init__(self,r):
        self.r = r

    def area(self):
        return 3.1416 * self.r * self.r

vals = input().split()

if len(vals) == 3:
    r1 = Rectangle(int(vals[1]),int(vals[2]))
    print(f"Area: {r1.area():.2f}")

else:
    c1 = Circle(int(vals[1]))
    print(f"Area: {c1.area():.2f}")
```

```python
def safe_divide(a,b):
    try:
        val = a/b
        print(val)
    except ZeroDivisionError:
        print("Division by zero")

nums = input().split()
safe_divide(int(nums[0]), int(nums[1]))
```
