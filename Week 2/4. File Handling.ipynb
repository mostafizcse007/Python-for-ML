##### Reading a file

```python
file = open('sample.txt','r')
content = file.readlines()
print(content)
file.close()
```

##### Read line by lines

```python
file = open('sample.txt','r')
content = file.readlines()
for i in content:
    print(i,end='')
file.close()
```

```python
with open('sample.txt','r') as file: # automatically close file
    for i in file:
        print(i.strip())
```

##### Writing a file

```python
with open('write.txt','w') as file:
    file.writelines('hello there\n'
                    'how are you\n'
                    'are you going well')
```

##### Appending a file

```python
with open('write.txt','a') as file:
    file.writelines('\na new line')
```

##### File pointer/cursor

```python
with open('sample.txt','r') as file:
    print(file.tell()) # initial cursor state
    print(file.read(4)) # telling the cursor where to start
    print(file.seek(10)) # moving the cursor without doing anything
    print(file.tell())
```

###### number of lines, words, characters, save to another file

```python
with open('test_1.txt','r') as file:
    lines =0
    words = 0
    chars = 0
    arr = []
    for txt in file:
        temp = txt.strip()
        lines += 1
        words += len(temp.split(' '))
        chars += len(txt.replace(' ',''))
        arr.append(txt)
print(lines,words,chars)
with open('test_2.txt','w') as file:
    file.writelines(arr)
```

##### Read and write inside one object

```python
with open('write_read.txt','w+') as file:
    file.writelines('Hard but not impossible\n'
                    'Every tensor is a matrix but not every matrix is a tensor')
    file.seek(0)
    print(file.read())
```

```python
with open('email.txt','r') as file:
    emails = file.readlines()
    emails = [x.lower() for x in emails]
    emails = list(set(emails))
    emails.sort()
    print(emails)
with open('unique_email.txt','w+') as file:
    file.writelines(emails)
```

```python
with open('names.txt','r') as file:
    names = file.read()
    names = names.split()
    names = [x.capitalize() for x in names]
    names = list(set(names))
    names.sort()
    print(names)
```

```python
with open('clean_names.txt', 'w+') as file:
    for x in names:
        file.write(x)
        file.write('\n')
```
