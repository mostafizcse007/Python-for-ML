######  Given two sets of friends from two people, find the mutual friends, unique friends of each, and the total number of unique friends.

```python
from markdown_it.rules_core.normalize import NULL_RE

a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}
print(f"Mutual friends: {a_friends & b_friends}")
print(f"Unique to a: {a_friends - b_friends}")
print(f"Unique to a: {b_friends - a_friends}")
print(f"Total unique friends: {len(a_friends|b_friends)}")
```

###### Sort Students by Marks (Tuple + Dictionary)


```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]
students.sort(key=lambda x:x[1],reverse=True)
print("Top 3 students:")
for i,j in students[0:3]:
    print(f"{i} - {j}")
```

###### Word Length Filter (Lambda + Filter)


```python
words = ["sun", "planet", "moon", "star", "universe"]
filtered_words = filter(lambda x: len(x)>4,words)
print(list(filtered_words))
```

###### Square of Even Numbers (Map + Filter + Lambda)

```python
nums = [1, 2, 3, 4, 5, 6]
new_nums = map(lambda x:x**2,filter(lambda x:x%2==0,nums))
arr = [x**2 for x in nums if x%2==0]
print(arr,list(new_nums))
```
