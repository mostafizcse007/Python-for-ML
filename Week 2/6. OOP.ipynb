```python
class Phone:
    category = "Electrical"
    def __init__(self,model,battery,camera):
        self.model = model
        self.battery = battery
        self.camera = camera
    def __str__(self):
        return f"{self.model} {self.battery} {self.camera}"

p1 = Phone('123',5000,64)

print(p1)
```

##### Inheritance

```python
class Smart(Phone):
    def __init__(self, model, battery, camera,no_of_folds):
        super().__init__(model, battery, camera)
        self.no_of_folds = no_of_folds

    def __str__(self):
        return f"{super().__str__()} {self.no_of_folds}"


p2 = Smart('Samsung',5000,108,3)
print(p2)

```

```python
class Father:
    age = 40
class Mother:
    hair_color = 'black'
class Son(Father,Mother):
    pass
s1 = Son()
print(s1.hair_color,s1.age)
```

##### Polymorphism

```python
class Addition:
    def add(self,*args):
        return sum(args)

a1 = Addition()
print(a1.add(1,2))
print(a1.add(1,2,3))
```

```python
class Camera:
    def __init__(self,name):
        self.name = name
    def capture(self):
        print(f'Photo captured by {self.name} at {self.resolution} MP')


class Mobile(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    def capture(self):
        super().capture()

class DSLR(Camera):
    def __init__(self, name,resolution):
        super().__init__(name)
        self.resolution = resolution

class Drone(Camera):
    def __init__(self, name,resolution):
        super().__init__(name)
        self.resolution = resolution

phone = Mobile('Samsung',100)
dslr = DSLR('Canon',1000)
drone = Drone('DJI',10000)
phone.capture()
dslr.capture()
drone.capture()
```
