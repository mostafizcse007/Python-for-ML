#### Iterators
###### an object that produces values one at a time when you ask for them.

```python
s = {1,2,3,4,5}
it = iter(s)
for i in range(5):
    print(next(it))
```

### Generators
###### A generator is an iterator that produces values lazily

```python
def data_loader(chunk_size,lst):
    for i in range(len(lst)):
        yield lst[i:i+chunk_size]
lst = list(range(500))
x = data_loader(10,lst)
print(x)
print(next(x))
```

#### Lambda

```python
# lambda parameter:extresssion(one expression only)
sq_add = lambda x,y: x**2 + y**2
print(sq_add(2,3))
print(sq_add(2,2))
```

### Map

```python
# lst = [1,2,3,4,5,6]
# arr = [x**2 for x in lst]

lst = [1,2,3,4,5,6]


def squared(x):
    return x**2
arr = list(map(squared,lst))
arr
```

#### Filter

```python
numbers = list(range(100))
even = filter(lambda x:x%2==0,numbers)
list(even)
```

##### Reduce

```python
from functools import reduce
```

```python
lst = [1,2,3,4,5,6]
summation = reduce(lambda x,y:x+y,lst)
summation
```

```python
lst = [1,2,3,4,5,6]
max_val = reduce(lambda x,y:x if x>y else y,lst)
max_val
```
